```python
from qwerty import sobrenomeNaOrdem
import unittest
```

```python
class NomeTeste(unittest.TestCase):
    def test_sobrenomeNaOrdem(self):
        nomeCompleto = sobrenomeNaOrdem('Joao','Madureira','Silva')
        self.assertEqual(nomeCompleto, 'Joao Silva Madureira')
    def test_sobrenomeNaOrdem2(self):
        nomeCompleto = sobrenomeNaOrdem('Madureira','Silva','Joao')
        self.assertEqual(nomeCompleto, 'Joao Silva Madureira)
```

```python
unittest.main(argv=[''], exit=False)
```

```python
class Casa():
    imobiliaria = "RONALDINHO BOSTE LEGAL"


    def __init__(self, rua , bairro , numero, CEP):
        self.rua = rua
        self.bairro = bairro
        self.CEP = CEP
        self.numero = numero
    def getimobiliaria(self):
        return self.imobiliaria
    def getrua(self):
        return self.rua
    def getbairro(self):
        return self.bairro
    def getCEP(self):
        return self.CEP
    def getnumero(self):
        return self.numero
    def setimobiliaria(self, i):
        self.imobiliaria = i
    def setrua(self, r):
        self.rua = r
    def setbairro(self, b):
        self.bairro = b
    def setCEP(self, c):
        self.CEP = c
    def setnumero(self, n):
        self.numero = n
    def enderecoCompleto(self):
        return f"Endereco Completo: {self.rua}, {self.numero}. {self.bairro} - CEP:{self.CEP}"
```

```python
casa1 = Casa(rua="Rua qwerty da point", numero="158", CEP="098765432", bairro="interlagos")
casa2 = Casa(rua="ronaldo", numero="ronaldinho", CEP="ronaldo boste", bairro="ronaldo legal")
```

```python
print(casa1.enderecoCompleto())
print(casa2.enderecoCompleto())
```

```python
print(casa1.imobiliaria)
print(casa2.imobiliaria)
```

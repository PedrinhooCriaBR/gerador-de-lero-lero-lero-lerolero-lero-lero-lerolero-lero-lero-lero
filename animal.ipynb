```python
class AnimalSelvagem():
    def mover(self):
        print("estou correndo")

    def oQueSou(self):
        print("cachorro")

    def come(self):
        print("Comendo")

class AnimalDomestico():
    def mover(self):
        print("estou andando")
    def getDono(self):
        return self.dono

class Cachorro(AnimalSelvagem, AnimalDomestico):
    def __init__(self, dono):
        self.dono = dono
        print("Cachorro criado")

    def late(self):
        print("auau!")
```

```python
c = Cachorro("Pedrinho é seu dono")
print(c.getDono())
```
